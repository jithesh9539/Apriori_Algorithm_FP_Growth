```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori 
```

```python
!pip install apyori
```

```python
data=pd.read_csv('/content/Market_Basket_Optimisation.csv',header=None)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
#For Apriori need to convert data into list format
transactions = []

for i in range(0,len(data)):
    transactions.append([str(data.values[i,j]) for j in range(0,20) if str(data.values[i,j])!='0'])

```

```python
transactions[1]
```

```python
associate_rules=apriori(transactions, min_support=0.003, min_confidance=0.2, min_lift=3, min_length=2)
associate_results=list(associate_rules)
```

```python
associate_results
```

```python
print(len(associate_results))
```

```python
print(associate_results[0])
```

```python
df_results = pd.DataFrame(associate_results)
```

```python
df_results.head()
```

```python
support = df_results.support
```

```python
first_values = []
second_values = []
third_values = []
fourth_value = []
```

```python
for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_value.append(single_list[3])
    
```

```python
lhs = pd.DataFrame(first_values)
rhs = pd.DataFrame(second_values)

confidance=pd.DataFrame(third_values,columns=['Confidence'])

lift=pd.DataFrame(fourth_value,columns=['lift'])
```

```python
df_final = pd.concat([lhs,rhs,support,confidance,lift], axis=1)
df_final
```

```python
df_final.fillna(value=' ', inplace=True)
df_final.head()
```

```python
df_final.columns = ['lhs',1,'rhs',2,3,'support','confidence','lift']
df_final.head()
```

```python
df_final['lhs'] = df_final['lhs'] + str(", ") + df_final[3]
```

```python
df_final['rhs'] = df_final['rhs']+str(", ")+df_final[2] + str(", ") + df_final[3]

```

```python
df_final.head()
```

```python
df_final['lhs'] = df_final['lhs'] + str(", ") + df_final[1]
```

```python
df_final['rhs'] = df_final['rhs']+str(", ")+df_final[2] + str(", ") + df_final[3]
```

```python
df_final.head()
```

```python
df_final.drop(columns=[1,2,3],inplace=True)
```

```python
df_final.head()
```
